```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas
from scipy import spatial

path = "lyrics_all.tsv"

data = pandas.read_csv(path, header = 0, delimiter = "\t", names = ['name', 'lyrics'])
```

```python
data
```

```python
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000)
```

```python
bow = vectorizer.fit_transform(data['lyrics'])
```

```python
bow_array = bow.toarray()
```

```python
bow_array
```

```python
1 - spatial.distance.cosine(bow_array[1], bow_array[5])
```

```python
len(bow_array)
```

```python
data['name'][0]
```

```python
l = len(bow_array)
m = 0
n = 0

# closest song to first:

for i in range(1,l):
    res = 1 - spatial.distance.cosine(bow_array[0], bow_array[i])
    if (res > m):
        m = res
        n = i
```

```python
print(m, data['name'][n])
```
