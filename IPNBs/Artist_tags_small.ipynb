## Теги last.fm небольшого количества исполнителей (240) 

```python
import json

path = '/mnt/hdd3/artist_small.json'


with open(path) as data_file:
    data_d = json.load(data_file)
```

```python
%pprint
```

```python
tags = []

for key in data_d:
    tmp_dict = data_d[key]
    for t_key in tmp_dict:
        tags.append(t_key)
```

```python
len(tags)
```

```python
tags.sort()
```

```python
from collections import Counter

tag_count = Counter(tags)
```

```python
tag_popular = {}

for key in tag_count:
    if (tag_count[key] > 9):
        tag_popular[key] = tag_count[key]
```

```python
import operator
print(sorted(tag_popular.items(), key=operator.itemgetter(1)))
```

```python
tag_list = []

for key in tag_popular:
    tag_list.append(key)
    
```

```python
tag_list
```

```python
l = len(tag_list)
```

```python
l_d = len(data_d)
```

```python
tags_2d = [[] for i in range(l_d)]

cnt = 0
for key in data_d:
    tmp_dict = data_d[key]
    tmp_list = []
    for t_key in tmp_dict:
        tmp_list.append(t_key)
        
    tags_2d[cnt] = tmp_list
    cnt += 1
```

```python
tmp = []
for j in range(l_d):
    if 'Bossa Nova' in tags_2d[j]:
        tmp.extend(tags_2d[j])
        
Counter(tmp)
```

```python
tmp = []
for j in range(l_d):
    if 'pop' in tags_2d[j]:
        tmp.extend(tags_2d[j])
        
Counter(tmp)
```

```python
joint_tags = [[] for i in range(l)]

for i in range(l):
    tmp = []
    for j in range(l_d):
        if tag_list[i] in tags_2d[j]:
            tmp.extend(tags_2d[j])
    joint_tags[i].extend(tmp)
```

```python
print(tag_list[0], "\n", Counter(joint_tags[0]))
```
