# Кластеризация knn

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import json
import pandas as pd
```

```python
f_1 = open("x.json")
f_2 = open("labels.json")

x = json.load(f_1)
lab = json.load(f_2)

x_new = []           # накосячила до этого с форматом, поэтому извращения
for line in x:
    line_new = json.loads(line)
    x_new.append(line_new)
    
    
X = np.asarray(x_new)
L = [int(l) for l in lab]
```

```python
neigh = KNeighborsClassifier(n_neighbors = 4)
neigh.fit(X, L)
```

```python
path = "vect.json"
data = json.load(open(path))
data_np = np.asarray(data)
pred = neigh.predict(data_np)
pr_unique = np.unique(pred)

idx = pd.read_csv("idx.txt", sep = "\t", header = None)
```

```python
for label in pr_unique:
    
    ids = np.where(label == pred)
    print([idx.values[i] for i in ids])
    print("\n")
```
